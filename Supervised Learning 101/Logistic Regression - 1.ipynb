```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
#We are considering only two species.
df = df[df["species"] != "virginica"]
```

```python
df["species"] = df["species"].map({"setosa":0, "versicolor":1})
```

```python
df.head()
```

```python
#Spliting dataset into dependant and independant variable
X = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)
```

```python
#Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
#Model is made, result can be obtained by fit and predict, but we need to do hyperparameter tuning for better/optimization of the model
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Giving parameters options to choose from, so that we can get best out of it.
params = {"penalty":['l1', 'l2', 'elasticnet'], "max_iter":[10, 20, 30, 40, 50, 75, 100, 200, 250, 300], "C":[1,2,3,4,5,10,20,30,40,50]}
```

```python
#Now, this will become our model
classifier_rgsr = GridSearchCV(classifier, param_grid=params,scoring="accuracy",cv=5)
```

```python
classifier_rgsr.fit(X_train, y_train)
```

```python
#Hence, best params are
print(classifier_rgsr.best_params_)
```

```python
print(classifier_rgsr.best_score_)
```

```python
#Predictions
y_pred = classifier_rgsr.predict(X_test)
```

```python
#Metrics
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

So, our model is completely right (1) so its very good (might be overfitting)
